```python

# Calculator using class and object
class calc():
     def add(a,b):
      return a+b
     def sub(a,b):
      return a-b

print(calc.sub(5,10))
print(calc.add(5,10))
```

```python
# Area using class
class area():
     def tri(l,b):
      return (1/2)*l*b
     def rect(l,b):
      return l*b
     def square(l,b):
      return l**2
print(area.tri(5,10))
print(area.rect(5,10))
print(area.square(5,10))
```
