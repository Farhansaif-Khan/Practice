```python
# Flip
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('download.webp')
if img is None:
  print("error : img not found")
else:
  cv2_imshow(img)
flp= cv2.flip(img,1)
cv2_imshow(flp)
```

```python
# Threshold
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('download.webp')
if img is None:
  print("error : img not found")
else:
  cv2_imshow(img)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
th_val,th_img=cv2.threshold(gray,60,255,cv2.THRESH_BINARY)

cv2_imshow(th_img)
print(th_val)
```

```python
# Contrast
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread('download.webp')
if img is None:
  print("error : img not found")
else:
  cv2_imshow(img)
r_min = np.min(img)
r_max = np.max(img)
contrast_stretched = (img - r_min) / (r_max - r_min)*255
contrast_stretched = contrast_stretched.astype(np.uint8)
cv2_imshow(contrast_stretched)
```
